```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("C:fmax_vs_Pe.dat")

Pe = data[:, 0]
f_max = data[:, 1]
```

```python
plt.plot(Pe, f_max, marker='o', linestyle='-')
plt.xlabel('Péclet Number (Pe)')
plt.ylabel('Average Maximum Cluster Size (<f_max>)')
plt.title('Average Maximum Cluster Size vs Péclet Number')
plt.grid()
plt.show()
```
