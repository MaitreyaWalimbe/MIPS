```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = r"C:/Users/Maitreya/mips_simulation/week1/task3/"
```

```python
phi_list = [0.2, 0.4, 0.6]
pe_list = [5.0, 20.0, 50.0, 100.0]
```

```python
for phi in phi_list:
    plt.figure(figsize=(8,6))

    phi_tag = int(phi*10)

    for Pe in pe_list:
        filename = f"msd_phi{phi_tag}_Pe{int(Pe)}.dat"
        full_path = path + filename

        data = np.loadtxt(full_path)

        t = data[:,0]
        msd = data[:,1]

        plt.plot(t, msd, label=f"Pe={Pe}")

    plt.xlabel("Time")
    plt.ylabel("Mean Squared Displacement (MSD)")
    plt.title(f"MSD vs Time for phi={phi}")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()
```
