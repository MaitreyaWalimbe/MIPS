```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t, x1, x2 = np.loadtxt("trajectory.dat").T
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x1, t, lw=1, label='Particle 1', color='blue')
plt.plot(x2, t, lw=1, label='Particle 2', color='green')
plt.xlabel('Position (x)')
plt.ylabel('Time (t)')
plt.title('Trajectories of Two Particles Over Time')
plt.legend()
plt.tight_layout()
plt.show()
```
