```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
traj_rat = np.loadtxt('C:/Users/Maitreya/mips_simulation/week2/task3/trajectory_rat.dat')
traj_srw = np.loadtxt('C:/Users/Maitreya/mips_simulation/week2/task3/trajectory_srw.dat')
```

```python
time_rat = traj_rat[:, 0]
positions_rat = traj_rat[:, 1:]

time_srw = traj_srw[:, 0]
positions_srw = traj_srw[:, 1:]
```

```python
plt.plot(positions_rat, time_rat, 'k.', ms=0.5)
plt.xlabel('Positions')
plt.ylabel('Time')
plt.title('Run and Tumble')
plt.tight_layout()
plt.show()
```

```python
plt.plot(positions_srw, time_srw, 'k.', ms=0.5)
plt.xlabel('Positions')
plt.ylabel('Time')
plt.title('Symmetric Random Walk')
plt.tight_layout()
plt.show()
```
